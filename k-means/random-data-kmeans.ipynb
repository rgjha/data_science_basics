```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import * 
# Uses Lloyd’s or Elkan’s algorithm
# C. Elkan. Using the Triangle Inequality to Accelerate k-Means. International Conference on Machine Learning, 2003.

%matplotlib inline
```

```python
X = -2 * np.random.rand(100,2) # Generate total set
X[50:100,:] = 1 + 2 * np.random.rand(50,2) # Assign second half to new set 
plt.scatter(X[:,0], X[:,1], s = 20, c = 'b')
plt.show()
```

```python
from sklearn.cluster import KMeans
Kmean = KMeans(n_clusters=2)
Kmean.fit(X)
Kmean.cluster_centers_
```

```python
plt.scatter(X[:,0], X[:,1], s =50, c='b')
plt.scatter(-0.99470835, -0.99549803, s=200, c='g', marker='s')
plt.scatter(1.91593745,  1.98597106, s=200, c='r', marker='s')
plt.show()
```
